## EV ANALYSIS

```python
# EV Distribution Across States

ev_data = ev_data[ev_data['State'] != 'Total']
ev_sorted = ev_data.sort_values(by='EV_Registrations', ascending=False)

fig = px.bar(ev_sorted, x='State', y='EV_Registrations', title='EV Registrations by State',)
fig.show()
```

#### Insight-
The visualization shows a highly uneven distribution of electric vehicle registrations across states. California dominates EV adoption by a very large margin, Washington is the second-highest, but still significantly behind California. Most states have relatively low EV registrations, showing that EV adoption is still concentrated in a few regions. The large gap between top states and others highlights regional disparity in EV adoption.

## Energy Data Preparation

```python
# Filtering energy data for year 2018
energy_2018 = energy_data[energy_data['Year'] == 2018]

# Aggregate total energy per state (all sources combined)
energy_state = energy_2018.groupby('State')['Energy_Generation'].sum().reset_index()

# Preview
energy_state.head()
```

```python
# Converting state codes to full names
energy_state = pd.merge(energy_state, state_codes, left_on='State', right_on='state_code', how='left')

# Replacing code with full name
energy_state['State'] = energy_state['state_name']

# Keeping required columns
energy_state = energy_state[['State', 'Energy_Generation']]

# Preview
energy_state.head()
```

```python
# Data Merging


merged_data = pd.merge(ev_data, energy_state, on='State', how='inner')
merged_data.head()
```

## Correlation Analysis

#### Correlation Between EV Registrations and Energy Generation

```python
fig = px.scatter(merged_data, x='Energy_Generation', y='EV_Registrations', hover_name='State', title='EV Registrations vs Energy Generation (2018)')
fig.show()
```

#### Insight-

- There is a general upward tendency: states with higher energy generation often have higher EV registrations, but the pattern is not strongly linear.
- A few states act as outliers:
  - One state (likely California) shows very high EV registrations relative to others.
  - Some states generate large amounts of energy but have comparatively low EV adoption.
- Most states are clustered at lower EV registration levels, even when energy generation varies widely.


```python
merged_data[['EV_Registrations', 'Energy_Generation']].corr()
```


#### Conclusion:
The correlation coefficient between EV registrations and energy generation is approximately **0.31**. While energy availability contributes to EV adoption, it is not the primary driving factor. A combination of policy support, infrastructure, and socio-economic conditions is required to accelerate EV growth.

## TIME SERIES ANALYSIS

```python
# Total energy per year
energy_yearly = energy_data.groupby('Year')['Energy_Generation'].sum().reset_index()

fig = px.line(energy_yearly, x='Year', y='Energy_Generation', title='Total Energy Generation Over Time')
fig.show()
```

#### Insight-

The line chart shows the trend of total energy generation from 1990 to 2019.

- There is a clear **long-term upward trend** from the early 1990s to the late 2000s, indicating sustained growth in energy production.
- Growth **slows and stabilizes after ~2008–2012**, with fluctuations around a relatively flat level.
- A noticeable **dip around 2008–2009** suggests the impact of external factors (e.g., economic slowdown).
- Post-2010, energy generation remains **stable with minor variations**, indicating a mature and balanced energy supply.

### SEASONAL TRENDS / LONG TERM SHIFTS

```python
# Focusing on major energy sources
major_sources = ['Coal', 'Natural Gas', 'Hydroelectric Conventional', 'Wind', 'Solar']

energy_filtered = energy_data[energy_data['Energy_Source'].isin(major_sources)]
energy_trend = energy_filtered.groupby(['Year', 'Energy_Source'])['Energy_Generation'].sum().reset_index()

fig = px.line(energy_trend, x='Year', y='Energy_Generation', color='Energy_Source', title='Energy Generation by Source Over Time')
fig.show()
```

#### Insight-

This chart shows how different energy sources have changed over the years.

- Coal was the main source of energy earlier, but its use is decreasing over time.
- Natural gas is increasing steadily and has become one of the major energy sources.
- Hydroelectric energy remains mostly stable with small changes.
- Wind energy is growing, especially after 2010, but still lower than major sources.

## GEOSPATIAL ANALYSIS

This chart shows the top 10 states with the highest number of EV registrations.

- California clearly leads with the highest EV registrations, far ahead of all other states.
- Washington is the second highest, but there is a large gap compared to California.

- The remaining states in the top 10 have relatively similar and lower values.
